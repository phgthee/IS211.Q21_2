```python
# Query 1: Tìm 5 nhân viên có doanh thu cao nhất ở cn1 và cn2
from py2neo import Graph

# Chi nhánh 1
graph_cn1 = Graph(
    "bolt://26.84.188.242:7687",
    auth=("neo4j", "123456789")
)

# Chi nhánh 2
graph_cn2 = Graph(
    "bolt://26.7.224.249:7687",
    auth=("neo4j", "phthyngx")
)

query = """
MATCH (nv:NhanVien)-[:TAO]->(hd:HoaDon)

RETURN nv.hoTen AS NhanVien,
       COUNT(hd) AS SoHoaDon,
       SUM(hd.tongTien) AS TongDoanhThu

ORDER BY TongDoanhThu DESC
LIMIT 5
"""

result_cn1 = graph_cn1.run(query)
result_cn2 = graph_cn2.run(query)

print("===== DOANH THU NHAN VIEN CN1 =====")

for row in result_cn1:

    print(
        row["NhanVien"],
        row["SoHoaDon"],
        row["TongDoanhThu"]
    )

print("===== DOANH THU NHAN VIEN CN2 =====")

for row in result_cn2:

    print(
        row["NhanVien"],
        row["SoHoaDon"],
        row["TongDoanhThu"]
    )
```

```python
# Query 2: Thống kê khách hàng có TongChiTieu > 5000000 (Khách VIP)
from py2neo import Graph

# Chi nhánh 1
graph_cn1 = Graph(
    "bolt://26.84.188.242:7687",
    auth=("neo4j", "123456789")
)

# Chi nhánh 2
graph_cn2 = Graph(
    "bolt://26.7.224.249:7687",
    auth=("neo4j", "phthyngx")
)

query = """
MATCH (kh:KhachHang)-[:DAT]->(hd:HoaDon)

WITH kh,
     SUM(hd.tongTien) AS TongChiTieu

WHERE TongChiTieu > 5000000

RETURN kh.hoTen AS KhachHang,
       TongChiTieu
ORDER BY TongChiTieu DESC
"""

result_cn1 = graph_cn1.run(query)
result_cn2 = graph_cn2.run(query)

print("===== KH VIP CN1 =====")

for row in result_cn1:

    print(
        row["KhachHang"],
        row["TongChiTieu"]
    )

print("===== KH VIP CN2 =====")

for row in result_cn2:

    print(
        row["KhachHang"],
        row["TongChiTieu"]
    )
```

```python
# Query 4: UNION sản phẩm của 2 chi nhánh và loại bỏ trùng
from py2neo import Graph

# Chi nhánh 1
graph_cn1 = Graph(
    "bolt://26.84.188.242:7687",
    auth=("neo4j", "123456789")
)

# Chi nhánh 2
graph_cn2 = Graph(
    "bolt://26.7.224.249:7687",
    auth=("neo4j", "phthyngx")
)

query = """
MATCH (sp:SanPham)

RETURN sp.maSanPham AS MaSP,
       sp.tenSanPham AS TenSP,
       sp.giaBan AS GiaBan
"""

cn1 = graph_cn1.run(query).data()
cn2 = graph_cn2.run(query).data()

tong = cn1 + cn2

unique = {}

for row in tong:

    unique[row["MaSP"]] = (
        row["TenSP"],
        row["GiaBan"]
    )

print("===== SAN PHAM TOAN HE THONG =====")

for masp, value in unique.items():

    print(
        masp,
        value[0],
        value[1]
    )
```

```python
# Query 5: Lọc sản phẩm có giá bán lớn hơn giá vốn 30%
from py2neo import Graph

# Chi nhánh 1
graph_cn1 = Graph(
    "bolt://26.84.188.242:7687",
    auth=("neo4j", "123456789")
)

# Chi nhánh 2
graph_cn2 = Graph(
    "bolt://26.7.224.249:7687",
    auth=("neo4j", "phthyngx")
)

query = """
MATCH (sp:SanPham)

WHERE sp.giaBan > sp.giaVon * 1.3

RETURN sp.tenSanPham AS SanPham,
       sp.giaVon AS GiaVon,
       sp.giaBan AS GiaBan
ORDER BY GiaBan DESC
"""

result_cn1 = graph_cn1.run(query)
result_cn2 = graph_cn2.run(query)

print("===== SAN PHAM LOI NHUAN CAO CN1 =====")

for row in result_cn1:

    print(
        row["SanPham"],
        row["GiaVon"],
        row["GiaBan"]
    )

print("===== SAN PHAM LOI NHUAN CAO CN2 =====")

for row in result_cn2:

    print(
        row["SanPham"],
        row["GiaVon"],
        row["GiaBan"]
    )
```

```python
# Query 6: Gom nhóm số lượng sản phẩm theo danh mục
from py2neo import Graph

# Chi nhánh 1
graph_cn1 = Graph(
    "bolt://26.84.188.242:7687",
    auth=("neo4j", "123456789")
)

# Chi nhánh 2
graph_cn2 = Graph(
    "bolt://26.7.224.249:7687",
    auth=("neo4j", "phthyngx")
)

query = """
MATCH (sp:SanPham)-[:THUOC_DANHMUC]->(dm:DanhMuc)

RETURN dm.tenDanhMuc AS DanhMuc,
       COUNT(sp) AS SoLuongSanPham

ORDER BY SoLuongSanPham DESC
"""

result_cn1 = graph_cn1.run(query)
result_cn2 = graph_cn2.run(query)

print("===== DANH MUC CN1 =====")

for row in result_cn1:

    print(
        row["DanhMuc"],
        row["SoLuongSanPham"]
    )

print("===== DANH MUC CN2 =====")

for row in result_cn2:

    print(
        row["DanhMuc"],
        row["SoLuongSanPham"]
    )
```
