```python
# Thêm CUSTOMER vào CN2 từ CN1
# Kết nối với CN2
# import thư viện
from py2neo import Graph

# Kết nối database của chi nhánh 1
#cn1 = Graph(
#    "bolt://26.84.188.242:7687",
#    auth=("neo4j", "123456789")
#)

# Kết nối database của chi nhánh 2
cn2 = Graph(
    "bolt://26.7.224.249:7687",
    auth=("neo4j", "phthyngx")
)

# Kiểm tra kết nối
try:

    result2 = cn2.run(
        "RETURN 1 AS test"
    ).data()

    #print("Kết nối CN1 thành công:", result1)
    print("Kết nối CN2 thành công:", result2)

except Exception as e:

    print("Kết nối thất bại:", e)
```

```python
# Thêm CUSTOMER từ CN1 sang CN2
query = """

CREATE (kh:KhachHang {

    maKhachHang:'KH999',
    hoTen:'Nguyen Van A',
    sdt:'0909999999',
    diemTichLuy:1000

})

"""

cn2.run(query)

print("Thêm CUSTOMER vào CN2 thành công")
```

```python
# Cập nhật CUSTOMER từ CN1 sang CN2
query = """

MATCH (kh:KhachHang {

    maKhachHang:'KH999'

})

SET kh.hoTen = 'Tran Thi Nga',
    kh.diemTichLuy = 5000

"""

cn2.run(query)

print("Cập nhật CUSTOMER tại CN2 thành công")
```

```python
# Xóa CUSTOMER từ CN1 sang CN2
query = """

MATCH (kh:KhachHang {

    maKhachHang:'KH999'

})

DETACH DELETE kh

"""

cn2.run(query)

print("Xóa CUSTOMER tại CN2 thành công")
```

```python
# Thêm DANH MỤC vào CN2 từ CN1
query = """

CREATE (d:DanhMuc {

    maDanhMuc:'DM999',
    tenDanhMuc:'Do Uong Co Con'

})

"""

cn2.run(query)

print("Thêm DANH MỤC vào CN2 thành công")
```

```python
# Cập nhật DANH MỤC từ CN1 sang CN2
query = """

MATCH (d:DanhMuc {

    maDanhMuc:'DM999'

})

SET d.tenDanhMuc = 'Do Uong Nhap Khau'

"""

cn2.run(query)

print("Cập nhật DANH MỤC tại CN2 thành công")
```

```python
# Xóa DANH MỤC ở CN2 từ CN1 
query = """

MATCH (d:DanhMuc {

    maDanhMuc:'DM999'

})

DETACH DELETE d

"""

cn2.run(query)

print("Xóa DANH MỤC tại CN2 thành công")
```
